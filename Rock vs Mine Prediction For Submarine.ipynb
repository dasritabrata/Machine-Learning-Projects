```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Data Collection and Processing

```python
sonar_data=pd.read_csv("/Users/ritabratadas/Desktop/Machine Learning/PROJECTS/DATASETS/Sonar data.csv",header=None)
```

### Separating data and labels

```python
X=sonar_data.drop(columns=60)
y=sonar_data[60]
```

```python
X.head()
```

```python
y.head()
```

## Training and Test Data

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=1,stratify=y)
```

```python
print((X.shape,X_train.shape,X_test.shape))
```

```python
X_train
```

## Model Training --> Logistic Regression

```python
model=LogisticRegression()
```

### Training the Logistic Regression Model with training data

```python
model.fit(X_train,y_train)
```

## Model Evaluation

### Accuracy on Training Data

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,y_train)
print(f"Accuracy on Training Data is {training_data_accuracy}")
```

### Accuracy on Test Data

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,y_test)
print(f"Accuracy on Test Data is {test_data_accuracy}")
```
